```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score

# Dataset
data = {
    'plot': [
        'A detective investigates a murder case',
        'Two lovers fall in love',
        'A superhero saves the city',
        'A haunted house scares a family',
        'Friends go on a funny road trip',
        'A warrior fights enemies',
        'Ghosts haunt a hospital',
        'A comedian creates funny situations',
        'An action hero fights terrorists',
        'A prince falls in love with a girl',
        'Monsters attack a town',
        'Police hunt a serial killer'
    ],
    'genre': [
        'Thriller',
        'Romance',
        'Action',
        'Horror',
        'Comedy',
        'Action',
        'Horror',
        'Comedy',
        'Action',
        'Romance',
        'Horror',
        'Thriller'
    ]
}

df = pd.DataFrame(data)

# Features and Labels
X = df['plot']
y = df['genre']

# Convert text to numbers
vectorizer = TfidfVectorizer()
X = vectorizer.fit_transform(X)

# Split dataset
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.25, random_state=42
)

# Train model
model = MultinomialNB()
model.fit(X_train, y_train)

# Test model
predictions = model.predict(X_test)

print("Accuracy:", accuracy_score(y_test, predictions))

# Predict new movie
new_movie = ["A detective investigates strange murders in a haunted mansion"]

new_movie_vector = vectorizer.transform(new_movie)

result = model.predict(new_movie_vector)

print("Predicted Genre:", result[0])
```
